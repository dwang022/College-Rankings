```python

# Create a scatter plot
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(
    x=df['overall_score'],
    y=df['employment_outcomes'],
    c='blue',  # Color of the markers
    alpha=0.7,  # Transparency
    edgecolors='black',  # Edge color of markers
)

# Add labels and title
ax.set_xlabel('Overall Score')
ax.set_ylabel('Employment Outcomes')
ax.set_title('Relationship between Overall Score and Employment Outcomes')

# Add a line of best fit (regression line)
slope, intercept = np.polyfit(df['overall_score'], df['employment_outcomes'], 1)
line_fit = np.polyval([slope, intercept], df['overall_score'])
ax.plot(df['overall_score'], line_fit, color='red', linestyle='--', label='Line of Best Fit')

# Calculate R-squared
residuals = df['employment_outcomes'] - line_fit
SS_res = np.sum(residuals**2)
SS_tot = np.sum((df['employment_outcomes'] - np.mean(df['employment_outcomes']))**2)
r_squared = 1 - (SS_res / SS_tot)

# Annotate plot with R-squared value
ax.annotate(f'R-squared = {r_squared:.2f}',
            xy=(0.03, 0.95),
            xycoords='axes fraction',
            fontsize=12,
            ha='left',
            va='top')


# Show the plot
plt.grid(True)
plt.show()
```

```python

# Aggregate data by region
region_data = full.groupby("region")[["overall_score"]].mean().reset_index()

# Set style
sns.set(style='whitegrid')

# Plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(region_data["region"], region_data["overall_score"], color="skyblue", edgecolor='black')  # Add edgecolor for better visibility
plt.xlabel("Region", fontsize=14)
plt.ylabel("Average Overall Score", fontsize=14)
plt.title("Average Overall Score by Geographic Region", fontsize=16)
plt.yticks(fontsize=12)  # Adjust font size of y-axis ticks
plt.grid(axis='y', linestyle='--', alpha=0.5)  # Add grid lines for better visualization

# Add data labels
for i, score in enumerate(region_data["overall_score"]):
    plt.text(i, score, f'{score:.2f}', ha='center', va='bottom', fontsize=10)

# Adjust layout
plt.tight_layout()

# Show the plot


plt.savefig("score_by_region.png")
plt.show()

```

```python
# Aggregate data by region
region_data2 = full.groupby("region")[["employment_outcomes"]].mean().reset_index()

# Set style
sns.set(style='whitegrid')

# Plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(region_data2["region"], region_data2["employment_outcomes"], color="skyblue", edgecolor = 'black')
plt.xlabel("Region")
plt.ylabel("Average Employment Outcomes")
plt.title("Average Employment Outcomes by Geographic Region")
plt.grid(axis='y', linestyle='--', alpha=0.5)  # Add grid lines for better visualization

# Add data labels
for i, score in enumerate(region_data2["employment_outcomes"]):
    plt.text(i, score, f'{score:.2f}', ha='center', va='bottom', fontsize=10)

# Adjust layout
plt.tight_layout()


plt.savefig("employment_outcomes_by_region.png")
plt.show()

```

```python
# Iterate over each numerical variable and create a bar chart
for column in full.select_dtypes(include=['float64', 'int64']).columns:
    # Aggregate data by region
    variable_data = full.groupby("region")[column].mean().reset_index()

    # Plot bar chart
    plt.figure(figsize=(10, 6))
    plt.bar(variable_data["region"], variable_data[column], color="skyblue")
    plt.xlabel("Region")
    plt.ylabel(f"Average {column}")
    plt.title(f"Average {column} by Geographic Region")
    plt.show()
```

```python
# Aggregate data by region
country_data = full.groupby("country")[["overall_score"]].mean().reset_index()

# Plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(country_data["country"], country_data["overall_score"], color="skyblue")
plt.xlabel("Country")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.ylabel("Average Overall Score")
plt.title("Average Overall Score by Country")

plt.savefig("score_by_country.png")
plt.show()

```

```python
# Aggregate data by region
numerical_cols = full.select_dtypes(include=['float64', 'int64']).columns.drop(["rank", "latitude", "longitude", "...1", "overall_score", "employment_outcomes", "international_faculty_ratio", "sustainability"])  # Exclude "rank" column

region_data = full.groupby("region")[numerical_cols].mean().reset_index()

# Plot stacked bar chart
plt.figure(figsize=(10, 6))
bottom = None
for col in numerical_cols:
    plt.bar(region_data["region"], region_data[col], label=col, bottom=bottom)
    if bottom is None:
        bottom = region_data[col]
    else:
        bottom += region_data[col]
plt.xlabel("Region")
plt.ylabel("Average Score")
plt.title("Stacked Bar Chart of Average Variables by Geographic Region")
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))


plt.savefig("stacked_bar_chart_by_region.png", bbox_inches='tight')
plt.show()



# consider dropping more cols
```

```python
def categorize_rank(rank):
    if rank <= 20:
        return "Top-ranked"
    elif rank <= 50:
        return "Average-ranked"
    else:
        return "Low-ranked"

full["performance_category"] = full["rank"].apply(categorize_rank)

# Calculate the proportion of universities in each performance category for each region
proportions = full.groupby(["region", "performance_category"]).size().unstack().fillna(0)

# Plot pie charts for each region and save them as PNGs
for region in proportions.index:
    non_zero_proportions = proportions.loc[region][proportions.loc[region] > 0]

    plt.figure(figsize=(8, 6))
    plt.pie(proportions.loc[region], labels=proportions.columns, autopct='%1.1f%%', startangle=140)
    plt.title(f"Proportion of Universities by Performance Category in {region}")
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
    
    # Save each plot as PNG
    # plt.savefig(f"proportion_{region}.png", bbox_inches='tight')
    
    # Show the plot (optional)
    plt.show()
    
    # Close the plot to release memory
    # plt.close()
```
